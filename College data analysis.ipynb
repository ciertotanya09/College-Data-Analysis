## Model Creation 

Importing kmeans from sklearn 

```python
from sklearn.cluster import KMeans
```

Initializing the Kmeans with 2 clusters.

We're taking n_cluster = 2 since we've to group it into 2 categories

```python
kmeans = KMeans(n_clusters=2,random_state=1)
```

```python
X = df.drop('Private',axis=1)
```

```python
kmeans.fit(X)
```

```python
kmeans.cluster_centers_
```

## Evaluation

```python
pred = kmeans.labels_
```

```python
from sklearn.preprocessing import LabelEncoder
enc=LabelEncoder()
df['Private']=enc.fit_transform(df['Private'])
```

Create a confusion matrix and classification report to see how well the Kmeans clustering worked without being given any labels

```python
from sklearn.metrics import confusion_matrix,classification_report
print(confusion_matrix(df['Private'],pred))
print(classification_report(df['Private'],pred))
```

Our model is doing pretty great job as it just find patterns in the data without actually looking at the labels.
